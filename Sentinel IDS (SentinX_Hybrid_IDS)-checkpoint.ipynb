```python
from scapy.all import sniff, IP, TCP, UDP

def analyze_packet(packet):
    # Check if the packet has an IP layer
    if packet.haslayer(IP):
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst
        proto = "TCP" if packet.haslayer(TCP) else "UDP" if packet.haslayer(UDP) else "Other"
        
        # Simple detection logic:
        # If traffic is coming from a specific IP (e.g., 192.168.1.1)
        print(f"[+] Traffic Detected: {src_ip} -> {dst_ip} | Protocol: {proto}")
        
        # NIDS Alert Example:
        if packet.haslayer(TCP) and packet[TCP].flags == "S":
            print(f"   [!] ALERT: Potential Port Scanning from {src_ip}")

print("Sniffing is starting... (10 packets will be captured)")

# Sniffing function: 'prn' specifies which function to run on each packet
# count=10 ensures the Jupyter cell stops automatically after 10 packets
sniff(prn=analyze_packet, count=10)

print("\nSniffing complete!")
```

```python
import os
import time
import threading
from scapy.all import sniff, IP
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler

# --- CONFIGURATION ---
TARGET_FOLDER = os.path.join(os.getcwd(), "SentinX_HIDS_Folder")
if not os.path.exists(TARGET_FOLDER):
    os.makedirs(TARGET_FOLDER)

# --- MODULE 1: HIDS (File Monitoring) ---
class HIDSHandler(FileSystemEventHandler):
    def on_any_event(self, event):
        if not event.is_directory:
            print(f"[HIDS ALERT] Activity on: {event.src_path} | Type: {event.event_type}")

def start_hids():
    event_handler = HIDSHandler()
    observer = Observer()
    observer.schedule(event_handler, TARGET_FOLDER, recursive=False)
    observer.start()
    print(f"[*] HIDS Active: Monitoring {TARGET_FOLDER}")
    try:
        while True: time.sleep(1)
    except: observer.stop()

# --- MODULE 2: NIDS (Network Monitoring) ---
def packet_callback(packet):
    if packet.haslayer(IP):
        # Alert if traffic is from a specific suspicious source (Example logic)
        print(f"[NIDS ALERT] Packet: {packet[IP].src} -> {packet[IP].dst}")

def start_nids():
    print("[*] NIDS Active: Sniffing Network Traffic...")
    sniff(prn=packet_callback, count=20) # 20 packets scan karega

# --- STARTING THE HYBRID SYSTEM ---
print("--- SentinX Hybrid IDS Starting ---")

# Threading taaki dono saath chalein
hids_thread = threading.Thread(target=start_hids, daemon=True)
hids_thread.start()

# NIDS start karein
start_nids()

print("--- Scan Complete ---")
```

```python
import os
import time
import pandas as pd
import matplotlib.pyplot as plt
from scapy.all import sniff, IP, TCP, UDP, ICMP
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
from datetime import datetime
import threading

# --- CONFIGURATION ---
LOG_FILE = "security_logs.txt"
WATCH_DIRECTORY = "./my_secure_folder"
NETWORK_PACKET_COUNT = 20

if not os.path.exists(WATCH_DIRECTORY):
    os.makedirs(WATCH_DIRECTORY)
    print(f"[!] Created directory to watch: {WATCH_DIRECTORY}")

# --- LOGGING ENGINE ---
def write_log(alert_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] [{alert_type}] {message}\n"
    with open(LOG_FILE, "a") as f:
        f.write(log_entry)
    print(log_entry.strip())

# --- MODULE 1: NETWORK SNIFFER ---
def analyze_packet(packet):
    if packet.haslayer(IP):
        src = packet[IP].src
        dst = packet[IP].dst
        proto = "Other"
        
        if packet.haslayer(TCP): proto = "TCP"
        elif packet.haslayer(UDP): proto = "UDP"
        elif packet.haslayer(ICMP): proto = "ICMP"

        # Basic anomaly logic: Flag specific IP ranges or protocols
        if proto == "ICMP":
            write_log("NETWORK_ALERT", f"ICMP (Ping) detected from {src} to {dst}")
        elif src.startswith("134.13"):
            write_log("NETWORK_CRITICAL", f"External traffic from suspicious range: {src}")

def start_network_monitor():
    print(f"[*] Starting Network Sniffer ({NETWORK_PACKET_COUNT} packets)...")
    sniff(prn=analyze_packet, count=NETWORK_PACKET_COUNT)
    print("[*] Network scan complete.")

# --- MODULE 2: FILE SYSTEM MONITOR ---
class HIDSHandler(FileSystemEventHandler):
    def on_modified(self, event):
        if not event.is_directory:
            write_log("FILE_MODIFIED", f"File altered: {os.path.basename(event.src_path)}")

    def on_created(self, event):
        write_log("FILE_CREATED", f"New file detected: {os.path.basename(event.src_path)}")

def start_file_monitor():
    event_handler = HIDSHandler()
    observer = Observer()
    observer.schedule(event_handler, WATCH_DIRECTORY, recursive=False)
    observer.start()
    print(f"[*] File Monitoring started on: {WATCH_DIRECTORY}")
    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()

# --- MODULE 3: VISUALIZATION ---
def generate_report():
    print("\n[*] Generating Security Summary...")
    if not os.path.exists(LOG_FILE):
        print("[-] No logs found to visualize.")
        return

    logs = []
    with open(LOG_FILE, "r") as f:
        for line in f:
            if "[" in line:
                # Extract alert type between brackets
                parts = line.split("] [")
                if len(parts) > 1:
                    alert_type = parts[1].split("]")[0]
                    logs.append(alert_type)

    if logs:
        df = pd.Series(logs).value_counts()
        df.plot(kind='bar', color=['red', 'orange', 'blue', 'green'])
        plt.title("Sentinel IDS: Intrusion Detection Summary")
        plt.xlabel("Alert Type")
        plt.ylabel("Frequency")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print("[-] Log file is empty.")

# --- MAIN EXECUTION ---
if __name__ == "__main__":
    # Start Network Monitoring in a separate thread
    net_thread = threading.Thread(target=start_network_monitor, daemon=True)
    net_thread.start()

    # Start File Monitoring (This will run until you manually stop it)
    print("[!] Sentinel Hybrid IDS is ACTIVE. Press Ctrl+C to stop and see report.")
    try:
        start_file_monitor()
    except KeyboardInterrupt:
        print("\n[!] Shutting down Sentinel...")
        generate_report()
```

```python
import os
import shutil
import time
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler

# --- CONFIGURATION ---
WATCH_DIR = "./secure_zone"
QUARANTINE_DIR = "./quarantine"
# Types of files we don't want strangers putting in our folder
BANNED_EXTENSIONS = ['.exe', '.sh', '.bat', '.py']

# Setup folders
for folder in [WATCH_DIR, QUARANTINE_DIR]:
    if not os.path.exists(folder):
        os.makedirs(folder)

class SentinelResponse(FileSystemEventHandler):
    def on_created(self, event):
        if event.is_directory:
            return

        file_name = os.path.basename(event.src_path)
        file_ext = os.path.splitext(file_name)[1].lower()

        print(f"\n[!] ALERT: New file detected: {file_name}")
        
        # ACTIVE RESPONSE: Quarantine suspicious files
        if file_ext in BANNED_EXTENSIONS:
            print(f"[!] ACTION: Suspicious extension {file_ext} detected!")
            self.quarantine_file(event.src_path, file_name)
        else:
            print(f"[*] LOG: Routine file '{file_name}' added to logs.")

    def quarantine_file(self, src_path, file_name):
        dest_path = os.path.join(QUARANTINE_DIR, file_name)
        try:
            # Move the file immediately
            shutil.move(src_path, dest_path)
            print(f"[SUCCESS] File moved to {QUARANTINE_DIR}. Potential threat neutralized.")
        except Exception as e:
            print(f"[ERROR] Could not quarantine file: {e}")

# --- MAIN INTERFACE ---
def run_sentinel():
    event_handler = SentinelResponse()
    observer = Observer()
    observer.schedule(event_handler, WATCH_DIR, recursive=False)
    
    print("="*40)
    print("      SENTINEL ACTIVE RESPONSE SYSTEM      ")
    print("="*40)
    print(f"[*] Monitoring: {WATCH_DIR}")
    print(f"[*] Quarantine: {QUARANTINE_DIR}")
    print("[*] Press Ctrl+C to shut down.")
    
    observer.start()
    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        observer.stop()
        print("\n[!] Sentinel shutting down safely.")
    observer.join()

if __name__ == "__main__":
    run_sentinel()
```

```python
import os
import shutil
import time
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler

# --- CONFIGURATION ---
WATCH_DIR = "./secure_zone"
QUARANTINE_DIR = "./quarantine"
BANNED_EXTENSIONS = ['.exe', '.sh', '.bat', '.py']

# --- EMAIL SETTINGS ---
SMTP_SERVER = "smtp.gmail.com"
SMTP_PORT = 587
SENDER_EMAIL = "raorituj125@gmail.com"       # your Gmail
SENDER_PASSWORD = "xddz boiy sgsg qpks"  # App Password
RECEIVER_EMAIL = "ritusrm11@gmail.com"      # alert recipient

def send_security_alert(file_name):
    """Sends an automated email alert when a threat is detected."""
    msg = MIMEMultipart()
    msg['From'] = SENDER_EMAIL
    msg['To'] = RECEIVER_EMAIL
    msg['Subject'] = "🚨 SECURITY ALERT: Sentinel IDS"

    body = f"""
    Warning! Sentinel IDS has detected a potential threat.
    
    Event: Suspicious File Detected
    File Name: {file_name}
    Action Taken: File moved to Quarantine.
    Timestamp: {time.ctime()}
    """
    msg.attach(MIMEText(body, 'plain'))   # ✅ attach body

    try:
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()                 # ✅ start TLS
        server.login(SENDER_EMAIL, SENDER_PASSWORD)
        server.send_message(msg)
        server.quit()
        print("📧 Email Alert sent successfully!")
    except Exception as e:
        print(f"❌ Email not sent: {e}")

class SecurityHandler(FileSystemEventHandler):
    def on_created(self, event):
        if not event.is_directory:
            file_ext = os.path.splitext(event.src_path)[1].lower()
            if file_ext in BANNED_EXTENSIONS:
                file_name = os.path.basename(event.src_path)
                quarantine_path = os.path.join(QUARANTINE_DIR, file_name)
                shutil.move(event.src_path, quarantine_path)
                print(f"[!] ALERT: Suspicious file moved to quarantine -> {file_name}")
                send_security_alert(file_name)

# --- Ensure directories exist ---
os.makedirs(WATCH_DIR, exist_ok=True)
os.makedirs(QUARANTINE_DIR, exist_ok=True)

# --- Start Monitoring ---
event_handler = SecurityHandler()
observer = Observer()
observer.schedule(event_handler, WATCH_DIR, recursive=False)

print(f"--- Sentinel IDS Active ---\nMonitoring folder: {WATCH_DIR}")
observer.start()

try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    observer.stop()
observer.join()
print("Monitoring stopped.")
```
